```python
import os
import cv2
import numpy as np
from PIL import Image, UnidentifiedImageError
from tqdm import tqdm
import rasterio
import shutil

dataset_root = "/content/EuroSAT_RGB"
output_dir = "/content/cleaned_dataset"
target_size = (64, 64)
filter_type = "median"
valid_exts = [".jpg", ".jpeg", ".png", ".tif", ".tiff"]

if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir)

def apply_filter(img):
    if filter_type == "gaussian":
        return cv2.GaussianBlur(img, (3, 3), 0)
    elif filter_type == "median":
        return cv2.medianBlur(img, 3)
    return img

def read_image_any_format(path):
    ext = os.path.splitext(path)[1].lower()
    if ext in [".tif", ".tiff"]:
        try:
            with rasterio.open(path) as src:
                arr = src.read()  # shape: (bands, height, width)
                if arr.shape[0] >= 3:
                    # Use bands 4,3,2 (common RGB order for Sentinel-2)
                    rgb = np.stack([
                        arr[3-1],  # band 3 index in python = 2
                        arr[2-1],
                        arr[1-1]
                    ], axis=-1)
                    rgb = cv2.normalize(rgb, None, 0, 255, cv2.NORM_MINMAX)
                    return rgb.astype(np.uint8)
                else:
                    return None
        except:
            return None
    else:
        return cv2.imread(path, cv2.IMREAD_UNCHANGED)

for root, dirs, files in os.walk(dataset_root):
    for file in tqdm(files, desc=f"Processing in {root}"):
        ext = os.path.splitext(file)[1].lower()
        if ext not in valid_exts:
            continue

        img_path = os.path.join(root, file)
        img = read_image_any_format(img_path)

        if img is None:
            print(f"⚠️ Skipping unreadable: {img_path}")
            continue

        if len(img.shape) == 2:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        elif img.shape[2] == 4:
            img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)

        img = cv2.resize(img, target_size)
        img = apply_filter(img)

        rel_path = os.path.relpath(root, dataset_root)
        save_folder = os.path.join(output_dir, rel_path)
        os.makedirs(save_folder, exist_ok=True)
        save_path = os.path.join(save_folder, os.path.splitext(file)[0] + ".png")
        cv2.imwrite(save_path, img)

print("✅ Data cleaning & preprocessing complete!")
```

```python
import os, glob
from collections import defaultdict

CLEAN_ROOT_CANDIDATE = "/content/cleaned_dataset"  # adjust if you saved elsewhere

def has_images(path):
    exts = ("*.png","*.jpg","*.jpeg","*.tif","*.tiff")
    return any(glob.glob(os.path.join(path, e)) for e in exts)

def count_images(path):
    total = 0
    for e in ("*.png","*.jpg","*.jpeg","*.tif","*.tiff"):
        total += len(glob.glob(os.path.join(path, e)))
    return total

def find_class_root(root):
    """
    Heuristic: return the first directory whose *immediate* subfolders
    contain images (>=5 subfolders with images). This typically is the
    level where subfolders are the 10 class names.
    """
    best = None
    for cur, dirs, files in os.walk(root):
        # Check only one level down
        child_dirs = [os.path.join(cur, d) for d in dirs]
        class_like = [d for d in child_dirs if has_images(d)]
        if len(class_like) >= 5:  # EuroSAT has 10 classes; 5 is a safe threshold
            best = cur
            break
    return best

def summarize_tree(root, depth=3):
    print(f"Scanning: {root}")
    for cur, dirs, files in os.walk(root):
        rel = os.path.relpath(cur, root)
        if rel == ".": rel = ""
        lvl = rel.count(os.sep)
        if lvl > depth:
            continue
        img_cnt = count_images(cur)
        print("  " * lvl + f"📁 {rel or os.path.basename(root)}  (imgs: {img_cnt})")
        # show top-level class folder counts
        if lvl == 1:
            for d in dirs:
                dpath = os.path.join(cur, d)
                print("  " * (lvl+1) + f"- {d} (imgs: {count_images(dpath)})")

# ---- Locate class-root and show counts ----
class_root = find_class_root(CLEAN_ROOT_CANDIDATE)
if class_root is None:
    print("❌ Could not find a class-root. Here’s a quick tree to debug:")
    summarize_tree(CLEAN_ROOT_CANDIDATE, depth=2)
else:
    print("✅ Detected class-root:", class_root)
    summarize_tree(class_root, depth=1)
```

```python
# Use the detected class-root
clean_dir = class_root

# (Re)run the feature extraction pipeline safely:
!pip install scikit-image scikit-learn --quiet

import os, glob, numpy as np, cv2
from skimage.feature import local_binary_pattern
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

max_per_class = 600
hist_bins = 16
lbp_P, lbp_R = 8, 1
np.random.seed(42)

def list_images_by_class(root):
    paths, labels = [], []
    if not root or not os.path.exists(root): return paths, labels
    for cls in sorted([d for d in os.listdir(root) if os.path.isdir(os.path.join(root, d))]):
        cls_dir = os.path.join(root, cls)
        imgs = []
        for ext in ("*.png","*.jpg","*.jpeg","*.tif","*.tiff"):
            imgs.extend(glob.glob(os.path.join(cls_dir, ext)))
        if not imgs:
            continue
        if max_per_class and len(imgs) > max_per_class:
            imgs = list(np.random.choice(imgs, max_per_class, replace=False))
        for p in imgs:
            paths.append(p); labels.append(cls)
    return paths, np.array(labels)

def rgb_histogram(img_bgr, bins=16):
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    feats = []
    for ch in range(3):
        h = cv2.calcHist([img],[ch],None,[bins],[0,256]).flatten()
        h /= (h.sum() + 1e-8)
        feats.append(h)
    return np.concatenate(feats)

def edge_density(img_bgr):
    g = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    e = cv2.Canny(g, 100, 200)
    return np.array([(e>0).mean()], dtype=np.float32)

def lbp_texture(img_bgr, P=8, R=1, uniform=True):
    g = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    method = 'uniform' if uniform else 'default'
    lbp = local_binary_pattern(g, P, R, method=method)
    n_bins = P*(P-1)+3 if uniform else 256
    hist, _ = np.histogram(lbp, bins=n_bins, range=(0, n_bins), density=True)
    return hist.astype(np.float32)

def extract_features(paths, labels):
    X, y = [], []
    for p, lab in zip(paths, labels):
        img = cv2.imread(p, cv2.IMREAD_COLOR)
        if img is None:
            continue
        img = cv2.resize(img, (64,64))
        feat = np.concatenate([rgb_histogram(img,hist_bins), edge_density(img), lbp_texture(img, lbp_P, lbp_R, True)])
        X.append(feat); y.append(lab)
    return np.array(X, dtype=np.float32), np.array(y)

def plot_2d(X, y, title, method="pca"):
    if len(X)==0:
        print(f"⚠️ Skipping plot: {title} (no samples)"); return
    scaler = StandardScaler(); Xs = scaler.fit_transform(X)
    if method=="tsne":
        reducer = TSNE(n_components=2, init="pca", learning_rate="auto",
                       perplexity=min(30, max(5, len(X)//50)), random_state=42)
    else:
        reducer = PCA(n_components=2, random_state=42)
    X2 = reducer.fit_transform(Xs)
    classes = sorted(list(set(y)))
    cmap = plt.cm.get_cmap('tab10', len(classes))
    plt.figure(figsize=(7.5,6))
    for i, cls in enumerate(classes):
        idx = (y==cls)
        plt.scatter(X2[idx,0], X2[idx,1], s=9, alpha=0.7, label=cls, c=np.array([cmap(i)]))
    plt.title(title); plt.legend(markerscale=2, fontsize=8, ncol=2, frameon=False); plt.grid(alpha=0.2); plt.show()

def quick_linear_separability_score(X, y, cv=3):
    if len(X)==0:
        print("⚠️ Skipping separability score (no samples)"); return None
    scaler = StandardScaler(); Xs = scaler.fit_transform(X)
    clf = LogisticRegression(max_iter=200, multi_class="auto", solver="lbfgs")
    scores = cross_val_score(clf, Xs, y, cv=cv, scoring="accuracy")
    print(f"LogReg (hand-crafted features) CV accuracy: {scores.mean():.3f} ± {scores.std():.3f}")
    return scores.mean(), scores.std()

# Run on CLEANED
clean_paths, clean_labels = list_images_by_class(clean_dir)
print(f"[CLEANED] images: {len(clean_paths)} | classes: {len(set(clean_labels))}")
X_clean, y_clean = extract_features(clean_paths, clean_labels)
print("Feature shape (clean):", X_clean.shape)

plot_2d(X_clean, y_clean, "PCA — Cleaned (RGB hist + edges + LBP)", method="pca")
plot_2d(X_clean, y_clean, "t‑SNE — Cleaned (RGB hist + edges + LBP)", method="tsne")
quick_linear_separability_score(X_clean, y_clean, cv=3)
```

```python
# =========================
# Task 3 — CNN From Scratch
# =========================

import os, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print("TensorFlow:", tf.__version__)
tf.random.set_seed(42)

# -----------------------------
# 1) Data: load & split (80/20)
# -----------------------------
DATA_DIR = "/content/cleaned_dataset/EuroSAT_MS/EuroSAT_MS"  # <- update if needed
IMG_SIZE = (64, 64)
BATCH_SIZE = 64
VAL_SPLIT = 0.2
SEED = 42

train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    labels='inferred',
    label_mode='int',
    validation_split=VAL_SPLIT,
    subset='training',
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    labels='inferred',
    label_mode='int',
    validation_split=VAL_SPLIT,
    subset='validation',
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
)

class_names = train_ds.class_names
num_classes = len(class_names)
print("Classes:", class_names)

# Prefetch for speed
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.shuffle(1024).cache().prefetch(AUTOTUNE)
val_ds   = val_ds.cache().prefetch(AUTOTUNE)

# -----------------------------
# 2) Model: CNN (no pretrained)
#    - >=3 conv layers
#    - ReLU activations
#    - BatchNorm
#    - MaxPool
#    - Dropout
#    - Dense + Softmax
# -----------------------------
inputs = keras.Input(shape=(*IMG_SIZE, 3))

# Normalize [0,255] -> [0,1]
x = layers.Rescaling(1./255)(inputs)

# Block 1
x = layers.Conv2D(32, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)
x = layers.Conv2D(32, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)
x = layers.MaxPooling2D()(x)
x = layers.Dropout(0.20)(x)

# Block 2
x = layers.Conv2D(64, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)
x = layers.Conv2D(64, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)
x = layers.MaxPooling2D()(x)
x = layers.Dropout(0.25)(x)

# Block 3 (third+ conv layer)
x = layers.Conv2D(128, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)
x = layers.MaxPooling2D()(x)
x = layers.Dropout(0.30)(x)

# Head
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)
x = layers.Dropout(0.40)(x)

outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs)
model.summary()

# -----------------------------
# 3) Compile & Train
# -----------------------------
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

EPOCHS = 20
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
)

# -----------------------------
# 4) Plot training curves
# -----------------------------
def plot_history(h):
    acc = h.history["accuracy"]; val_acc = h.history["val_accuracy"]
    loss = h.history["loss"];     val_loss = h.history["val_loss"]

    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(acc, label="train acc")
    plt.plot(val_acc, label="val acc")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy"); plt.legend(); plt.grid(alpha=.2)

    plt.subplot(1,2,2)
    plt.plot(loss, label="train loss")
    plt.plot(val_loss, label="val loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss"); plt.legend(); plt.grid(alpha=.2)
    plt.show()

plot_history(history)

# Optionally save for later tasks
model.save("/content/cnn_baseline.keras")
print("Saved model to /content/cnn_baseline.keras")
```

```python
# =========================
# Save predictions to CSV (validation + optional test)
# =========================
import os, glob, math, numpy as np, pandas as pd, tensorflow as tf

ALLOWED_EXTS = (".png", ".jpg", ".jpeg")

def list_sorted_images_per_class(root, class_names):
    files_by_class = {}
    for cls in class_names:
        cls_dir = os.path.join(root, cls)
        imgs = []
        for ext in ("*.png","*.jpg","*.jpeg"):
            imgs += glob.glob(os.path.join(cls_dir, ext))
        imgs = sorted(imgs)  # Keras uses alphanumeric sorting
        files_by_class[cls] = imgs
    return files_by_class

def split_train_val(files, val_split=0.2):
    """Return (train_files, val_files) using the same rule as Keras:
       take the first (1 - split) for train, last split for val (per class)."""
    n = len(files)
    split_idx = int(math.floor(n * (1.0 - val_split)))
    return files[:split_idx], files[split_idx:]

def make_ds_from_paths(paths, img_size):
    """Create a tf.data.Dataset from file paths, preserving order (no shuffle)."""
    def load_img(p):
        img = tf.io.read_file(p)
        img = tf.io.decode_image(img, channels=3, expand_animations=False)
        img = tf.image.resize(img, img_size)
        # Model has a Rescaling layer, so keep raw [0..255] uint8/float; we’ll cast to float32 (no /255)
        img = tf.cast(img, tf.float32)
        return img
    path_ds = tf.data.Dataset.from_tensor_slices(paths)
    ds = path_ds.map(load_img, num_parallel_calls=tf.data.AUTOTUNE).batch(64).prefetch(tf.data.AUTOTUNE)
    return ds

# ---------- Build the exact validation file list ----------
files_by_class = list_sorted_images_per_class(DATA_DIR, class_names)

val_filepaths = []
for cls in class_names:
    train_files, val_files = split_train_val(files_by_class[cls], VAL_SPLIT)
    val_filepaths.extend(val_files)

# Sanity
print(f"Validation files detected: {len(val_filepaths)}")

# Build dataset in the SAME order as val_filepaths
val_pred_ds = make_ds_from_paths(val_filepaths, IMG_SIZE)

# Predict
probs = model.predict(val_pred_ds, verbose=1)
pred_indices = np.argmax(probs, axis=1)
pred_labels  = [class_names[i] for i in pred_indices]

# Save CSV: image_id = filename only
val_image_ids = [os.path.basename(p) for p in val_filepaths]
df_val = pd.DataFrame({"image_id": val_image_ids, "predicted_label": pred_labels})
val_csv_path = "/content/val_predictions.csv"
df_val.to_csv(val_csv_path, index=False)
print(f"✅ Saved validation predictions to: {val_csv_path} (rows: {len(df_val)})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# --- 1) Plot curves with best-epoch marker ---
hist = history.history
epochs = np.arange(1, len(hist["loss"]) + 1)
best_idx = int(np.argmin(hist["val_loss"]))  # 0-based

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(epochs, hist["accuracy"], label="train acc")
plt.plot(epochs, hist["val_accuracy"], label="val acc")
plt.axvline(best_idx+1, ls="--", alpha=.5, label=f"best val loss (ep {best_idx+1})")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy"); plt.legend(); plt.grid(alpha=.2)

plt.subplot(1,2,2)
plt.plot(epochs, hist["loss"], label="train loss")
plt.plot(epochs, hist["val_loss"], label="val loss")
plt.axvline(best_idx+1, ls="--", alpha=.5, label=f"best val loss (ep {best_idx+1})")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss"); plt.legend(); plt.grid(alpha=.2)
plt.show()

# --- 2) Evaluate generalization gap ---
train_acc_last = hist["accuracy"][-1]
val_acc_last   = hist["val_accuracy"][-1]
gap_last = train_acc_last - val_acc_last

train_acc_best = hist["accuracy"][best_idx]
val_acc_best   = hist["val_accuracy"][best_idx]
gap_best = train_acc_best - val_acc_best

print(f"Generalization gap (last epoch): {gap_last:.3f}")
print(f"Generalization gap (best val-loss epoch #{best_idx+1}): {gap_best:.3f}")
print(f"Train acc (last/best): {train_acc_last:.3f}/{train_acc_best:.3f}")
print(f"Val   acc (last/best): {val_acc_last:.3f}/{val_acc_best:.3f}")

# --- 3) Simple rule-based diagnosis ---
def diagnose(train_acc, val_acc, gap):
    if train_acc < 0.75 and val_acc < 0.70 and abs(gap) < 0.05:
        return "Likely UNDERFITTING (both train/val low, small gap)."
    if train_acc >= 0.90 and val_acc < 0.80 and gap >= 0.08:
        return "Likely OVERFITTING (train high, val much lower, big gap)."
    if train_acc >= 0.85 and val_acc >= 0.80 and gap < 0.06:
        return "Good fit / balanced (no major over/underfitting)."
    if val_acc > train_acc + 0.02:
        return "Possible regularization too strong or noisy labels (val > train)."
    return "Borderline case—check curves and confusion matrix."

print("\nDiagnosis (last epoch):", diagnose(train_acc_last, val_acc_last, gap_last))
print("Diagnosis (best val-loss epoch):", diagnose(train_acc_best, val_acc_best, gap_best))

# --- 4) Precise evaluation on train/val now ---
train_loss, train_acc_eval = model.evaluate(train_ds, verbose=0)
val_loss,   val_acc_eval   = model.evaluate(val_ds,   verbose=0)
print(f"\nEvaluate() -> Train acc: {train_acc_eval:.3f}, Val acc: {val_acc_eval:.3f}, Gap: {train_acc_eval - val_acc_eval:.3f}")

# --- 5) Confusion matrix + report on validation set ---
y_true = np.concatenate([y.numpy() for _, y in val_ds], axis=0)
y_prob = model.predict(val_ds, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("\nClassification report (validation):")
print(classification_report(y_true, y_pred, target_names=class_names, digits=3))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(7,6))
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix (Validation)")
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45, ha='right')
plt.yticks(tick_marks, class_names)
plt.ylabel('True label'); plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()
```

```python
# =========================
# Task 4 — Improved CNN end-to-end (Fixed LR logger)
# =========================
import os, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from sklearn.metrics import classification_report, confusion_matrix

print("GPU:", tf.config.list_physical_devices('GPU'))
tf.random.set_seed(42)

DATA_DIR = "/content/cleaned_dataset/EuroSAT_MS/EuroSAT_MS"  # change if needed
IMG_SIZE   = (64, 64)
BATCH_SIZE = 64
VAL_SPLIT  = 0.20
SEED       = 42

train_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR, validation_split=VAL_SPLIT, subset="training", seed=SEED,
    image_size=IMG_SIZE, batch_size=BATCH_SIZE, label_mode="int"
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR, validation_split=VAL_SPLIT, subset="validation", seed=SEED,
    image_size=IMG_SIZE, batch_size=BATCH_SIZE, label_mode="int"
)

class_names = train_ds.class_names
num_classes = len(class_names)
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.shuffle(2048).cache().prefetch(AUTOTUNE)
val_ds   = val_ds.cache().prefetch(AUTOTUNE)

# Data augmentation
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.10),
    layers.RandomContrast(0.10),
], name="augmentation")

# Model
def build_improved_model(img_size=IMG_SIZE, n_classes=num_classes, wd=1e-4, drop=0.35):
    inputs = keras.Input(shape=(*img_size, 3))
    x = layers.Rescaling(1./255)(inputs)
    x = data_augmentation(x)

    # Conv block 1
    x = layers.Conv2D(48, 3, padding="same", use_bias=False, kernel_regularizer=regularizers.l2(wd))(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x = layers.Conv2D(48, 3, padding="same", use_bias=False, kernel_regularizer=regularizers.l2(wd))(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x = layers.MaxPooling2D()(x); x = layers.Dropout(drop-0.10)(x)

    # Conv block 2
    x = layers.Conv2D(96, 3, padding="same", use_bias=False, kernel_regularizer=regularizers.l2(wd))(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x = layers.Conv2D(96, 3, padding="same", use_bias=False, kernel_regularizer=regularizers.l2(wd))(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x = layers.MaxPooling2D()(x); x = layers.Dropout(drop)(x)

    # Conv block 3
    x = layers.Conv2D(192, 3, padding="same", use_bias=False, kernel_regularizer=regularizers.l2(wd))(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x = layers.MaxPooling2D()(x); x = layers.Dropout(drop+0.05)(x)

    # Head
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(192, use_bias=False, kernel_regularizer=regularizers.l2(wd))(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
    x = layers.Dropout(drop+0.10)(x)

    outputs = layers.Dense(n_classes, activation="softmax")(x)
    return keras.Model(inputs, outputs, name="cnn_improved")

model = build_improved_model()
model.summary()

# Compile
optimizer = keras.optimizers.Adam(learning_rate=1e-3)
model.compile(optimizer=optimizer,
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

# Callbacks
cb_early = keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
cb_ckpt  = keras.callbacks.ModelCheckpoint("/content/cnn_improved_best.keras", monitor="val_loss", save_best_only=True)
cb_plateau = keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-5, verbose=1)

class LrLogger(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        lr = float(tf.keras.backend.get_value(self.model.optimizer.learning_rate))
        print(f"Epoch {epoch+1}: lr={lr:.6f}")

EPOCHS = 30
history = model.fit(
    train_ds, validation_data=val_ds, epochs=EPOCHS,
    callbacks=[cb_early, cb_ckpt, cb_plateau, LrLogger()],
    verbose=1
)

# Plot
def plot_history(h):
    acc = h.history["accuracy"]; val_acc = h.history["val_accuracy"]
    loss = h.history["loss"];     val_loss = h.history["val_loss"]
    epochs = range(1, len(acc)+1)
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1); plt.plot(epochs, acc, label="train acc"); plt.plot(epochs, val_acc, label="val acc")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy"); plt.legend(); plt.grid(alpha=.2)
    plt.subplot(1,2,2); plt.plot(epochs, loss, label="train loss"); plt.plot(epochs, val_loss, label="val loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss"); plt.legend(); plt.grid(alpha=.2)
    plt.show()
plot_history(history)

# Eval
val_loss, val_acc = model.evaluate(val_ds, verbose=0)
print(f"[Improved] Val accuracy: {val_acc:.4f} | Val loss: {val_loss:.4f}")

# Confusion matrix
y_true = np.concatenate([y.numpy() for _, y in val_ds], axis=0)
y_prob = model.predict(val_ds, verbose=0)
y_pred = np.argmax(y_prob, axis=1)
print("\nClassification report (validation):")
print(classification_report(y_true, y_pred, target_names=class_names, digits=3))
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(7,6))
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix (Validation)")
plt.colorbar()
ticks = np.arange(len(class_names))
plt.xticks(ticks, class_names, rotation=45, ha='right')
plt.yticks(ticks, class_names)
plt.ylabel('True'); plt.xlabel('Predicted')
plt.tight_layout(); plt.show()

model.save("/content/cnn_improved_final.keras")
print("Saved improved model to /content/cnn_improved_final.keras")
```

```python
import tensorflow as tf, numpy as np, matplotlib.pyplot as plt

def find_last_conv(model):
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name
    raise ValueError("No Conv2D layer found.")

def grad_cam(model, image_tensor, class_index=None, conv_layer_name=None):
    # image_tensor: shape (1, H, W, 3), raw (model has internal Rescaling)
    if conv_layer_name is None:
        conv_layer_name = find_last_conv(model)
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(image_tensor)
        if class_index is None:
            class_index = tf.argmax(preds[0])
        class_channel = preds[:, class_index]
    grads = tape.gradient(class_channel, conv_out)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_out = conv_out[0]
    heatmap = tf.reduce_sum(tf.multiply(pooled_grads, conv_out), axis=-1).numpy()
    heatmap = np.maximum(heatmap, 0)
    if heatmap.max() > 0: heatmap /= heatmap.max()
    return heatmap, int(class_index.numpy() if hasattr(class_index, 'numpy') else class_index)

def show_gradcam_on_image(model, img, class_names, true_label=None):
    x = tf.cast(tf.expand_dims(img, 0), tf.float32)  # (1,H,W,3)
    heatmap, pred_idx = grad_cam(model, x)
    pred_name = class_names[pred_idx]
    # resize heatmap to image
    heatmap_resized = tf.image.resize(heatmap[..., None], img.shape[:2]).numpy().squeeze()
    # overlay
    plt.figure(figsize=(8,3))
    plt.subplot(1,3,1); plt.imshow(img.numpy().astype(np.uint8)); plt.title("Image"); plt.axis("off")
    plt.subplot(1,3,2); plt.imshow(heatmap_resized, cmap="jet"); plt.title("Grad‑CAM"); plt.axis("off")
    plt.subplot(1,3,3);
    plt.imshow(img.numpy().astype(np.uint8));
    plt.imshow(heatmap_resized, cmap="jet", alpha=0.4)
    t = f"Pred: {pred_name}"
    if true_label is not None: t += f"\nTrue: {class_names[int(true_label)]}"
    plt.title(t); plt.axis("off")
    plt.tight_layout(); plt.show()

# show Grad‑CAM on a handful of val images
for imgs, labels in val_ds.take(1):
    for i in range(6):
        show_gradcam_on_image(model, imgs[i], class_names, true_label=labels[i])
```
